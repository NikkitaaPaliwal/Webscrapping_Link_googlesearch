```
#importing of libraries
from googlesearch import search
import nltk
import pandas as pd
nltk.download('punkt')

#input
ip=input("What would you like to search for? ")


#you will get here related links of the given input
extract_links = []
for url in search(ip):
     extract_links.append(url)
print(extract_links)
```

```
#links in form of dataframe
links_1 = pd.DataFrame(extract_links)
links_1.to_csv('generated_text.csv', index=True)
links_1
```

```
#extraction and output in form of table
import time
start = time.time()
import pandas as pd
import numpy as np
from newspaper import Article
links = extract_links
articles_info = []
for i in links:
    article_dict = {}
    article_dict["link"] = i
    art = Article(i)
    art.download()
    
    
    try:
        art.parse()
        article_dict["title"] = art.title
        article_dict["text"] = art.text
        art.nlp()
        article_dict["summary"] = art.summary
        article_dict["keywords"] = art.keywords
    except:
        article_dict["title"] = np.nan
        article_dict["text"] = np.nan
        article_dict["summary"] = np.nan
        article_dict["keywords"] = np.nan
        
        
    articles_info.append(article_dict)

corpus = pd.DataFrame(articles_info)
corpus_1 = corpus.dropna()
corpus_1.to_csv ("dataset.csv", index = False, header=True)
```
